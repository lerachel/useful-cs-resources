```python
import scipy.io as sio
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
# Load USPS.mat file

image_data = sio.loadmat('USPS.mat')
```

```python
# Extract matrix of images using key 'A' from dict image_data
A = image_data['A']
```

```python
print('Shape of matrix A:', A.shape)
```

```python
# Step 1: Normalize data

A_norm = normalize(A)
```

```python
# Step 2: Create a list of 5 PCA objects with different # components
# Number of components = 10, 50, 100, 150, 200

pca = []
components = [10, 50, 100, 150, 200]
for i in range(0,5):
    x = PCA(n_components=components[i])
    pca.append(x)
```

```python
# Step 3: Create a list of projections of A using the above PCA objects

A_pca = []
for i in range(0,5):
    x = pca[i].fit_transform(A_norm)
    A_pca.append(x)
```

```python
# Step 3: Create a list of reconstructed data from each projection of A

reconstruct = []
for i in range(0,5):
    x = pca[i].inverse_transform(A_pca[i])
    reconstruct.append(x)
```

```python
# Plot last two images from Original data and Reconstructed Data

# Last second image 
fig, ax = plt.subplots(1,6, figsize=(20,60))
fig.tight_layout()

ax[0].imshow(np.reshape(A[-2,],(16,16)))
ax[0].set_title('Original Image', fontweight="bold", size=24)

ax[1].imshow(np.reshape(reconstruct[4][-2,], (16,16)))
ax[1].set_title('PCA = 200', fontweight="bold", size=24)

ax[2].imshow(np.reshape(reconstruct[3][-2,], (16,16)))
ax[2].set_title('PCA = 150', fontweight="bold", size=24)

ax[3].imshow(np.reshape(reconstruct[2][-2,], (16,16)))
ax[3].set_title('PCA = 100', fontweight="bold", size=24)

ax[4].imshow(np.reshape(reconstruct[1][-2,], (16,16)))
ax[4].set_title('PCA = 50', fontweight="bold", size=24)

ax[5].imshow(np.reshape(reconstruct[0][-2,], (16,16)))
ax[5].set_title('PCA = 10', fontweight="bold", size=24)

# Last image
fig1, ax1 = plt.subplots(1,6, figsize=(20,60))
fig1.tight_layout()

ax1[0].imshow(np.reshape(A[-1,],(16,16)))
ax1[0].set_title('Original Image', fontweight="bold", size=24)

ax1[1].imshow(np.reshape(reconstruct[4][-1,], (16,16)))
ax1[1].set_title('PCA = 200', fontweight="bold", size=24)

ax1[2].imshow(np.reshape(reconstruct[3][-1,], (16,16)))
ax1[2].set_title('PCA = 150', fontweight="bold", size=24)

ax1[3].imshow(np.reshape(reconstruct[2][-1,], (16,16)))
ax1[3].set_title('PCA = 100', fontweight="bold", size=24)

ax1[4].imshow(np.reshape(reconstruct[1][-1,], (16,16)))
ax1[4].set_title('PCA = 50', fontweight="bold", size=24)

ax1[5].imshow(np.reshape(reconstruct[0][-1,], (16,16)))
ax1[5].set_title('PCA = 10', fontweight="bold", size=24)
```

```python
# Calculate reconstruction error using MSE - Mean Squared Error
err = []

for x in reconstruct:
    error = mean_squared_error(A, x)
    err.append(error)
```

```python
# Convert MSE and components arrays to list

err = list(err)
components = list(components)
```

```python
for i, x in zip(components, err):
    print("MSE of reconstructed data with %d components is %f" % (i, x))
    
print("\n* Smaller MSE means reconstructed data are closer to original data")

print("* Reconstructed data with higher # components have lower MSE") 
print("\nConclusion:")
print("Reconstructed data with more components will be more similar to original data.")
print("Higher number of components represent original data more accurately.")
```
