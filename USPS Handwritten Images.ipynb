```python
import scipy.io as sio
import numpy as np
from sklearn.decomposition import PCA
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
image_data = sio.loadmat('USPS.mat')
```

```python
A = image_data['A']
```

```python
print(A.shape)
```

```python
pca = []
components = [10, 50, 100, 150, 200]
for i in range(0,5):
    x = PCA(n_components=components[i])
    pca.append(x)
```

```python
A_pca = []
for i in range(0,5):
    x = pca[i].fit_transform(A)
    A_pca.append(x)
```

```python
reconstruct = []

for i in range(0,5):
    x = pca[i].inverse_transform(A_pca[i])
    reconstruct.append(x)
```

```python
# Second image 
fig, ax = plt.subplots(1,6, figsize=(20,60))
fig.tight_layout()

ax[0].imshow(np.reshape(A[2,],(16,16)))
ax[0].set_title('Original Image', fontweight="bold", size=24)

ax[1].imshow(np.reshape(reconstruct[4][2,], (16,16)))
ax[1].set_title('PCA = 200', fontweight="bold", size=24)

ax[2].imshow(np.reshape(reconstruct[3][2,], (16,16)))
ax[2].set_title('PCA = 150', fontweight="bold", size=24)

ax[3].imshow(np.reshape(reconstruct[2][2,], (16,16)))
ax[3].set_title('PCA = 100', fontweight="bold", size=24)

ax[4].imshow(np.reshape(reconstruct[1][2,], (16,16)))
ax[4].set_title('PCA = 50', fontweight="bold", size=24)

ax[5].imshow(np.reshape(reconstruct[0][2,], (16,16)))
ax[5].set_title('PCA = 10', fontweight="bold", size=24)

# 1000th image
fig1000, ax1000 = plt.subplots(1,6, figsize=(20,60))
fig1000.tight_layout()

ax1000[0].imshow(np.reshape(A[1000,],(16,16)))
ax1000[0].set_title('Original Image', fontweight="bold", size=24)

ax1000[1].imshow(np.reshape(reconstruct[4][1000,], (16,16)))
ax1000[1].set_title('PCA = 200', fontweight="bold", size=24)

ax1000[2].imshow(np.reshape(reconstruct[3][1000,], (16,16)))
ax1000[2].set_title('PCA = 150', fontweight="bold", size=24)

ax1000[3].imshow(np.reshape(reconstruct[2][1000,], (16,16)))
ax1000[3].set_title('PCA = 100', fontweight="bold", size=24)

ax1000[4].imshow(np.reshape(reconstruct[1][1000,], (16,16)))
ax1000[4].set_title('PCA = 50', fontweight="bold", size=24)

ax1000[5].imshow(np.reshape(reconstruct[0][1000,], (16,16)))
ax1000[5].set_title('PCA = 10', fontweight="bold", size=24)
```
